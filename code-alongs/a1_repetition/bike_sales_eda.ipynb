## länder - profit

```python
df[["Country", "Profit"]].groupby(["Country"]).mean(numeric_only=True).sort_values(
    by="Profit", ascending=False
).reset_index()
```

```python
df[["Country", "Profit"]].groupby(["Country"]).sum(numeric_only=True).sort_values(
    by="Profit", ascending=False
).reset_index()
```
